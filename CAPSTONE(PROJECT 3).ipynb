```python
# a.Build a simple linear model where dependent variable is ‘MonthlyCharges’ and independent variable is ‘tenure’
#    i.Divide the dataset into train and test sets in 70:30 ratio. 
#    ii.Build the model on train set and predict the values on test set
#    iii.After predicting the values, find the root mean square error
#    iv.Find out the error in prediction & store the result in ‘error’
#    v.Find the root mean square error
```

```python
x = pd.DataFrame(df['tenure'])
y = pd.DataFrame(df.MonthlyCharges)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=5,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X = train_x,
       y = train_y)
```

```python
y_pred = lr.predict(test_x)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(test_y,y_pred)
mse
```

```python
rsme = np.sqrt(mse)
rsme
```

```python
error = rsme
```

```python
error
```

```python
x = pd.DataFrame(df.loc[:,['tenure','MonthlyCharges']])
                        
y = pd.DataFrame(df['Churn'])
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.3,random_state=5)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=150).fit(X= train_x,
                                  y=train_y)
```

```python
y_pred = rfc.predict(test_x)
```

```python
confusion_matrix(test_y,y_pred)
```

```python
accuracy_score(test_y,y_pred)
```
